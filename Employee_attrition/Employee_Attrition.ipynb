```python
df=df.drop(['Over18','EmployeeCount','StandardHours','EmployeeNumber'],axis=1)
```

### Exploratory Data Analysis and Visualization

```python
df.hist(figsize=(15,15))
plt.tight_layout()
plt.show()
```

```python
# Group the Age in bins of 11
labels = ["{0} - {1}".format(i, i + 10) for i in range(18,60+1,11)]

df['Age_group']=pd.cut(df.Age, range(18,72,11), right=False, labels=labels)
```

```python
df.Age_group.value_counts()
```

```python
pd.crosstab(df.Age_group,df.Attrition).plot(kind='bar')
```

```python
df['Attrition'].value_counts().plot(kind='bar')
```
